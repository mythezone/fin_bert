```python
from data.process.bucketizer import UniformBucketizer
from data.process.bucketizer import ExponentialBucketizer
import matplotlib.pyplot as plt
import numpy as np 

sample = [x * 0.0001 for x in range(-1000, 1000)]


```

```python
for i in range(1,5):
    ub = UniformBucketizer(k=5+i*2)
    y = ub.encode(sample)
    plt.plot(sample, y)
```

```python
for i in range(10):

    eb = ExponentialBucketizer(k=20, e=1.1+i*0.1, zero=0.001)
    y = eb.encode(sample)
    y2 = ub.encode(sample)
    plt.plot(sample, y)
# plt.xscale('log')
# plt.xlim(-0.001, 0.001)
```

```python
for i in range(20):

    eb = ExponentialBucketizer(k=5+i, e=1.3, zero=0.001)
    y = eb.encode(sample)
    y2 = ub.encode(sample)
    plt.plot(sample, y)
```
